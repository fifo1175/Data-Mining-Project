```python
import json
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_excel("Denver_Crime_2018.xlsx")

offenseTypeOccurences = df['OFFENSE_TYPE_ID'].value_counts()
offenseCategoryOccurences = df['OFFENSE_CATEGORY_ID'].value_counts()

top5Type = offenseTypeOccurences.nlargest(5)
top5Category = offenseCategoryOccurences.nlargest(5)
print("Most common offense types:")
print(str(top5Type))
print("Most common offense categories:")
print(str(top5Category))

# bar chart for top 5 types
types = ["theft-items-from-vehicle", "theft-of-motor-vehicle", "theft-other", "criminal-trespassing", "theft-parts-from-vehicle"]
occurences = top5Type.tolist()

fig = plt.figure(figsize=(10,5))

plt.bar(types, occurences, color='blue', width=0.4)
plt.xlabel('Crime Types')
plt.ylabel('Number of Occurences')
plt.title('5 Most Common Types of Crime in Denver 2018')
plt.show()

# bar chart for top 5 categories
categories = ['all-other-crimes', 'larceny', 'public-disorder', 'theft-from-motor-vehicle', 'drug-alcohol']
occurences2 = top5Category.tolist()

fig2 = plt.figure(figsize=(10,5))

plt.bar(categories, occurences2, color='green', width=0.4)
plt.xlabel('Crime Categories')
plt.ylabel('Number of Occurences')
plt.title('5 Most Common Categories of Crime in Denver 2018')
plt.show()

categories = top5Category.tolist()





```
