```python

import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import math
import scipy as sp
import statistics as stats
import datetime
```

```python
def visualize_corr(dfW, dfC, year, month):

    dfC['counter'] = 1
    dfCCount = dfC.groupby(pd.Grouper(key='FIRST_OCCURRENCE_DATE', axis=0, freq='D')).sum()
    dfCCount = dfCCount.head(dfCCount.shape[0]-1)

    
    dfWCount = dfW.groupby('Date')['Temperature'].mean().to_frame().reset_index()
    dfWCount = dfWCount.head(dfWCount.shape[0]-1)

    ax = plt.gca()
    plt.title("Crime and Weather Correlation in "+month+' '+year)
    plt.xlabel("Date")
    ax.plot(dfWCount['Date'],dfCCount["counter"],color='red',linewidth=.8,label="Number of Crimes") 
    ax.plot(dfWCount['Date'],dfWCount["Temperature"],color='blue',linewidth=.8, label="Average Weather") 
    plt.setp(ax.get_xticklabels(), rotation=90) 
    ax.legend()
    plt.show()

```

```python
######
#2017#
######

dfC2017_05 = pd.read_excel('Denver Crime Data/2017/2017_May.xlsx')
dfW2017_05 = pd.read_csv('weather_data/denver_weather_data/all_weather_2017/05_2017.csv')

visualize_corr(dfW2017_05, dfC2017_05, "2017", "May")
```
