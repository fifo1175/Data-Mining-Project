```python
###
#compare monthly mean to the crime on a federal holiday
###

#list of all federal holiday dates (MONTH,DAY)
holidays_2017 = ['2017-01-02', '2017-01-16', '2017-02-20', '2017-05-29', '2017-07-04', '2017-09-04', '2017-11-23', '2017-11-24', '2017-12-25']
holidays_2018 = ['2018-01-01', '2018-01-15', '2018-02-19', '2018-05-28', '2018-07-04', '2018-09-03', '2018-11-22', '2018-11-23', '2018-12-25']
holidays_2019 = ['2019-01-01', '2019-01-21', '2019-02-18', '2019-05-27', '2019-07-04', '2019-09-02', '2019-11-28', '2019-11-29', '2019-12-25']
holidays_2020 = ['2020-01-01', '2020-01-20', '2020-02-17', '2020-05-25', '2020-07-03', '2020-07-04', '2020-09-07', '2020-11-26', '2020-11-27', '2020-12-25']
#holidays_2021 = [(1, 1), (1,18), (2,15), (5,31), (6,18), (7,4), (9,6), (11,25), (11,26), (12,24), (12,25)]



dates2017 = pd.date_range('2017-01-01','2017-12-31',freq='d').tolist()
dates2017.pop(0)

dates2018 = pd.date_range('2018-01-01','2018-12-31',freq='d').tolist()
#dates2018.pop(0)

dates2019 = pd.date_range('2019-01-01','2019-12-31',freq='d').tolist()
#dates2019.pop(0)

dates2020 = pd.date_range('2020-01-01','2020-12-31',freq='d').tolist()
#dates2020.pop(0)




#plot holidays and crime for each year
def crit_dates_plot(dfC, hol_arr, year_str, dates):
    #dfWCount = dfW.groupby('Date')['Temperature'].mean().to_frame().reset_index()

    dfC['counter'] = 1
    dfCCount = dfC.groupby(pd.Grouper(key='FIRST_OCCURRENCE_DATE', axis=0, freq='D')).sum()

    ax = plt.gca()
    plt.title("Crime in " + year_str + " With Federal Holidays Labeled")
    plt.xlabel("Date")

    ax.plot(dates,dfCCount["counter"],color='red',linewidth=.8,label="Crimes Per Day") 

    for ind, element in enumerate(dates):
        for i in hol_arr:
            date_string = element.date().strftime('%Y-%m-%d')
            if date_string == i:
                ax.plot(element.date(), dfCCount['counter'][ind], marker="o", markersize=10, markeredgecolor="black", markerfacecolor="green")


    ax.xaxis.set_major_locator(mdates.MonthLocator(bymonth=range(1,13,1)))
    plt.setp(ax.get_xticklabels(), rotation=90) 
    ax.legend()
    plt.show()



```

```python
crit_dates_plot(pd.read_excel('Denver Crime Data/2017/Denver_Crime_2017.xlsx'), holidays_2017, '2017', dates2017) #2017
crit_dates_plot(pd.read_excel('Denver Crime Data/2018/Denver_Crime_2018.xlsx'), holidays_2018, '2018', dates2018) #2018
crit_dates_plot(pd.read_excel('Denver Crime Data/2019/Denver_Crime_2019.xlsx'), holidays_2019, '2019', dates2019) #2019
crit_dates_plot(pd.read_excel('Denver Crime Data/2020/Denver_Crime_2020.xlsx'), holidays_2020, '2020', dates2020) #2020
```
