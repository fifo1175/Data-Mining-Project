```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
from sklearn import linear_model
```

```python
dfhouse = pd.read_csv('../../../Housing_Data/CHECKPOINT_1/crime_avgs.csv')
# read in house averages
dfcrime = pd.read_csv('../../../Housing_Data/CHECKPOINT_1/house_avgs.csv')

dfhouse = pd.read_csv("house_avgs.csv")
dfcrime = pd.read_csv("crime_avgs.csv")

cchouse = dfhouse.loc[dfhouse['Neighborhood']=='cherry-creek']

cpwhouse = dfhouse.loc[dfhouse['Neighborhood']=='city-park-west']

wwhouse = dfhouse.loc[dfhouse['Neighborhood']=='westwood']
cccrime = dfcrime.loc[dfcrime['Neighborhood']=='cherry-creek']
cpwcrime =dfcrime.loc[dfcrime['Neighborhood']=='city-park-west']
wwcrime=dfcrime.loc[dfcrime['Neighborhood']=='westwood']
```

```python
#CHERRY CREEK SCATTER PLOT 2017
x = np.linspace(1,12,12)
y12=cccrime.loc[cccrime['Year']=='2017'].iloc[:,-12:].values[0]
y11 = cchouse.loc[cchouse['Year']=='2017'].iloc[:,-12:].values[0]
fig, ax1 = plt.subplots()

ax2 = ax1.twinx()

# ax1.plot(x, y11, color="b",linestyle='dashed')
ax1.scatter(x,y11,c="b")
ax2.scatter(x, y12, color="r")

ax1.set_xlabel('Months')
ax1.set_ylabel('House prices', color="b")
ax2.set_ylabel('number of crimes', color="r")
ax1.set_ylim(min(y11),max(y11))
ax2.set_ylim(min(y12),max(y12))
plt.title("scatter plot for house prices and crimes over months in 2017 in cherry creek")


plt.show()
```

```python
#CHERRY CREEK REGRESSION 2017

# model = linear_model.LinearRegression()
# model.fit(np.reshape(y11, (len(y11),1)), np.reshape(y12, (len(y12),1)))
# fig, ax = plt.subplots()
# plt.plot(y11, model.predict(np.reshape(y11, (len(y11),1))), color='y')

# ax.set_xlim(min(y11),max(y11))
# ax.set_xlabel("house prices")
# ax.set_ylabel("Number of crimes")
# plt.show()
reg = linear_model.LinearRegression() 
e = [[e1] for e1 in y11]

# Fit the model using the lists 
reg.fit(e, y12) 

# Print out the linear regression line 
print(reg.coef_)
print(reg.intercept_)
x= np.linspace(1700000,1800000)
y = reg.coef_*x + reg.intercept_
plt.plot(x,y)
plt.title("Regression line between houses and crimes in 2017 in cherry creek")
plt.xlabel("house prices")
plt.ylabel("crimes")
plt.show()
```
